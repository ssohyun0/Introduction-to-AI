```python
import pandas as pd
import requests
from io import BytesIO

# 4. 웹 주소_xlsx
file_url = "https://raw.githubusercontent.com/MyungKyuYi/AI-class/refs/heads/main/combined_dataset-1.xlsx"

# 파일 다운로드
response = requests.get(file_url)
response.raise_for_status()

# 엑셀 파일을 데이터프레임으로 변환
df = pd.read_excel(BytesIO(response.content))

# 데이터프레임 확인
df.head()

######설명#######
#✅ CSV 파일 (.csv)
#CSV는 텍스트 기반(UTF-8, ASCII 등) 파일.
#pandas에서 바로 pd.read_csv("URL")로 읽을 수 있음.
#requests.get() 없이 바로 불러와도 문제 없음.

#✅ Excel 파일 (.xlsx)
#Excel 파일은 바이너리(Binary) 형식으로 저장.
#일반적인 텍스트 데이터처럼 바로 읽을 수 없음.
#반드시 파일을 다운로드 후, 변환해서 읽어야 함.
```
